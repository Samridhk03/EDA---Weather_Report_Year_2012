Exploratory Data Analysis for the Weather Report of the Year 2012

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/weather_2012.csv', parse_dates = True, index_col= 'Date/Time')
df
```

```python
df.info()
```

```python
df.shape
```

1. How did the temperature vary amongst the months during the year?

```python
#Grouping the data on the basis of month

monthly_data = df.groupby(df.index.month).mean()
monthly_data
```

```python
x = monthly_data.index
y = monthly_data['Temp (C)']
```

```python
plt.plot(x,y)
plt.title('Temperature Trend of Year 2012')
plt.xlabel('Months')
plt.xticks(x)
plt.ylabel('Temp in "C"')
plt.show()
```

```python
# Converting month numbers to actual month name in the x axis
import calendar
months = calendar.month_name[1:]
print(months)
```

```python
#Writing the code again to plot the chart

plt.plot(months,y)
plt.title('Temperature Trend of Year 2012')
plt.xlabel('Months')
plt.xticks(months, rotation = 50)
plt.ylabel('Temp in "C"')
plt.figure(figsize = (10,8))
plt.show()
```

2. Find the variation in mean visibility, aggregated by days of the week

```python
weekly_data = df.groupby(df.index.dayofweek).mean()
weekly_data
```

```python
plt.bar(weekly_data.index,weekly_data['Visibility (km)'])
plt.title('Visibility by the Week')
plt.xlabel('Days')
plt.xticks(weekly_data.index, calendar.day_abbr, rotation = 10)
plt.ylabel('Visibility (KM)')
```

3. Viewing the average Wind speed amongst the Months

```python
plt.plot(months, monthly_data['Wind Spd (km/h)'])
plt.title('Wind Speed during the Year 2012')
plt.xlabel('Months')
plt.ylabel('Wind Speed Km/Hr')
plt.xticks(months, rotation = 40)
plt.show()
```

4. Comparison between the mean of Dew and Humidity amongst the months

```python
fig, weather = plt.subplots()
weather.plot(months,monthly_data['Dew Point Temp (C)'], 'b.-')
weather.plot(months,monthly_data['Rel Hum (%)'], 'g--')
plt.xticks(months, rotation = 45)
fig.show()
```

5. Comparing the Overall Weather

```python
fig, overall_weather= plt.subplots()
overall_weather.plot(months, y, c = 'r')
overall_weather.plot(months, monthly_data['Wind Spd (km/h)'], 'b--')
overall_weather.plot(months,monthly_data['Dew Point Temp (C)'], 'g.-')
plt.xticks(months, rotation = 45)
fig.show()
```
